What is the relationship between bearing_trees and 60kgs_bag? How *strong* is this relationship? Use pearson's correlation coefficient.

Answer here:

All signs in pearsons correlation coefficient point to a strong correlation between bearing_trees and 60kgs_bag.

```python
x = coffee_df['bearing_trees']
y = coffee_df['60kgs_bag']
correlation_matrix = np.corrcoef(x, y)[ 0,1]
print(correlation_matrix)
```
